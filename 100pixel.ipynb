```python
from PIL import Image, ImageOps

# Open the image using the available file
image = Image.open("/content/WhatsApp Image 2026-07-21 at 23.12.13 (1).jpeg")

# Add a black border of 100 pixels
bordered_image = ImageOps.expand(
    image,
    border=100,
    fill="black"
)

# Save the output image
bordered_image.save("image_with_black_border.jpg")

print("Black border added successfully!")

# Display the image (optional)
bordered_image.show()
```

```python
from PIL import Image

# Open the image
image = Image.open("WhatsApp Image 2026-07-21 at 23.12.13 (1) (1).jpeg")

# Get image dimensions
width, height = image.size

# Crop 100 pixels from each side
cropped_image = image.crop((
    100,          # Left
    100,          # Top
    width - 100,  # Right
    height - 100  # Bottom
))

# Save the cropped image
cropped_image.save("cropped_image.jpg")

print("Image cropped successfully!")

# Display the image (optional)
cropped_image.show()
```
